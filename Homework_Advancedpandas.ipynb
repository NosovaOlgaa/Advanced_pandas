```python
'''Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. 
Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. 
Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.'''
```

```python
ratings = pd.read_csv('C:\\Users\\User\\Desktop\\Аналитик данных\\Python\\Файлы с лекций\\ratings.csv')
ratings.head()
```

```python
pt=ratings.pivot_table(index = 'userId', columns = 'rating', values = 'timestamp', aggfunc = 'count', fill_value = 0, 
                    margins = True)
pt.head()
```

```python
def watcher(param):
    """Киноман"""
    if param <= 100:
        return param == False
```

```python
pt['watcher'] = pt['All'].apply(watcher)

pt.loc[pt.watcher.isnull(), 'watcher']=pt['All']
pt.head()
```

```python
cinephile = pt.loc[pt.watcher != False]
cinephile.head()
```

```python
life_max=ratings.groupby('userId').max().reset_index()
life_min=ratings.groupby('userId').min().reset_index()

life_max.head()
```

```python
life_max_timestamp=life_max[['userId','timestamp']]
life_min_timestamp=life_min[['userId','timestamp']]

life_max_timestamp.head()
```

```python
life_max_timestamp.rename(columns={'timestamp': 'timestamp_max'}, inplace=True)
life_min_timestamp.rename(columns={'timestamp': 'timestamp_min'}, inplace=True)

#Что ему не нравится?
```

```python
life_average_timestamp=cinephile.merge(life_max_timestamp, on=["userId"], how='inner').merge(life_min_timestamp,on=["userId"], how='inner')
life_average_timestamp.head()
```

```python
def average(row):
    mean_=((row['timestamp_max']-row['timestamp_min'])//3600)
    return mean_
```

```python
life_average_timestamp['avg']=life_average_timestamp.apply(average, axis=1)

life_average_timestamp.head(10)
```

```python
'''Задание 4

Дана статистика услуг перевозок клиентов компании по типам (см. файл “Python_13_join.ipynb” в разделе 
Материалы для лекции «Продвинутый pandas» ---- Ноутбуки к лекции «Продвинутый pandas»).

Необходимо сформировать две таблицы:

*таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
*аналогичную таблицу по типам выручки с указанием адреса клиента

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.'''
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
#таблица с тремя типами выручки для каждого client_id без указания адреса клиента

revenue_all=rzd.merge(auto, on=["client_id"], how='outer').merge(air,on=["client_id"], how='outer')
revenue_all
```

```python
#таблица по типам выручки с указанием адреса клиента

revenue_all_with_address=revenue_all.merge(client_base, on=["client_id"], how='outer')
revenue_all_with_address
```
